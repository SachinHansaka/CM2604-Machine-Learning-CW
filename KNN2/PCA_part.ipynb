```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
url = 'spambase.data.csv'
#url = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"
spam = pd.read_csv(url, header=None)
X = spam.iloc[:, :-1].values
y = spam.iloc[:, -1].values
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Perform PCA
pca = PCA(n_components=2) # choose number of principal components to keep
X_pca = pca.fit_transform(X_scaled)
```

```python
print(X_scaled.shape)
print(X_pca.shape)
```

```python
# Create a new dataframe with the principal components and the target variable
df = pd.DataFrame(data = X_pca, columns = ['PC1', 'PC2'])
df['target'] = y
```

```python
# Visualize the results
sns.scatterplot(x='PC1', y='PC2', hue='target', data=df)
plt.show()
```
